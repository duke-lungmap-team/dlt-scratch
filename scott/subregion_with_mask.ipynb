```python
import os
import glob
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig_size = (4, 4)

img_file = "data/mouse_16.5_20x/LMEX0000000073_acta2_sftpc_sox9/2015-04-029_20X_C57Bl6_E16.5_LMM.14.24.4.46_SOX9_SFTPC_ACTA2_001.tif"
img = cv2.imread(img_file)
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
fig = plt.figure(figsize=fig_size)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB))
```

```python
c = np.array([[715.0, 706.0],
 [741.0, 724.0],
 [774.0, 731.0],
 [791.0, 776.0],
 [786.0, 813.0],
 [765.0, 835.0],
 [737.0, 828.0],
 [706.0, 818.0],
 [682.0, 798.0],
 [674.0, 767.0],
 [673.0, 744.0],
 [684.0, 724.0]], dtype='int')
```

```python
c
```

```python
b_rect = cv2.boundingRect(c)
```

```python
test_mask = np.zeros(img.shape[0:2], dtype=np.uint8)
```

```python
test_mask.shape
```

```python
cv2.drawContours(test_mask, [c], 0, 255, cv2.FILLED)
```

```python
fig = plt.figure(figsize=fig_size)
plt.imshow(cv2.cvtColor(test_mask, cv2.COLOR_GRAY2RGB))
```

```python
b_rect
```

```python
x1 = b_rect[0]
x2 = b_rect[0] + b_rect[2]
y1 = b_rect[1]
y2 = b_rect[1] + b_rect[3]
```

```python
x1
```

```python
fig = plt.figure(figsize=fig_size)
plt.imshow(cv2.cvtColor(test_mask[y1:y2, x1:x2], cv2.COLOR_GRAY2RGB))
```

```python
mask_img = cv2.bitwise_and(img, img, mask=test_mask)
plt.imshow(cv2.cvtColor(mask_img[y1:y2, x1:x2, :], cv2.COLOR_HSV2RGB))
```
