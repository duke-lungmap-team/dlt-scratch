```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
import json

%matplotlib inline
```

```python
def plot_rectangles(hsv_img, fig_size=(8, 8), rectangles=None):
    plt.figure(figsize=fig_size)
    ax = plt.gca() 
    plt.imshow(cv2.cvtColor(hsv_img, cv2.COLOR_HSV2RGB))
    
    if rectangles is not None:
        for rect in rectangles:
            ax.add_patch(
                patches.Rectangle(
                    xy=(rect['x1'], rect['y1']),
                    width=rect['x2'] - rect['x1'],
                    height=rect['y2'] - rect['y1'],
                    fill=False,
                    edgecolor='#00ff00'
                )
            )
```

```python
base_dir = "data/mouse_16.5_20x"

img_filename = "2015-04-029_20X_C57Bl6_E16.5_LMM.14.24.4.46_SOX9_SFTPC_ACTA2_001.tif"
img_path = "/".join([base_dir, img_filename])

json_filename = "distal_acinar_tubule.json"
json_path = "/".join([base_dir, json_filename])
```

```python
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
json_data = json.load(open(json_path))
```

```python
json_data[0].keys()
```

```python
for obj in json_data:
    if obj['image_path'] == img_filename:
        rectangles = obj['rects']
```

```python
print len(rectangles)
print(rectangles[0])
```

```python
plot_rectangles(img, rectangles=rectangles, fig_size=(16, 16))
```
