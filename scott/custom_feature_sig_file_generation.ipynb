```python
import os
import glob
import cv2
import pandas as pd

from lung_map_utils import utils

%matplotlib inline
```

```python
fig_size = (4, 4)
color_blur_kernel = (27, 27)
lrg_struct_blur_kernel = (91, 91)

METADATA_FILE = "/home/swhite/git/image-subregion-classifier/data/metadata_table_all.csv"

img_file = "data/mouse_16.5_20x/LMEX0000000073_acta2_sftpc_sox9/2015-04-029_20X_C57Bl6_E16.5_LMM.14.24.4.46_SOX9_SFTPC_ACTA2_001.tif"
img = cv2.imread(img_file)
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
source_file_name = os.path.basename(img_file)

# load metadata
metadata = pd.read_csv(METADATA_FILE, index_col=0)

# get metadata from file name
species, development, magnification, probes = utils.get_image_metadata(metadata, source_file_name)

probe_str = "_".join(sorted(probes))

train_dir = "/".join(
    [
        '/home/swhite/Projects/lung_map/custom_features/data',
        species,
        development,
        magnification,
        probe_str
    ]
)

class_paths = glob.glob("/".join([train_dir, "*"]))
class_map = {}

train_images = []

for i, class_path in enumerate(sorted(class_paths)):
    folder, class_name = os.path.split(class_path)

    class_id = i + 1
    class_map[class_id] = class_name

    train_images.extend(glob.glob("/".join([class_path, "*.tif"])))
```

```python
for sub_img_path in train_images:
    sub_img = cv2.imread(sub_img_path)
    try:
        sub_img = cv2.cvtColor(sub_img, cv2.COLOR_BGR2HSV)
    except:
        print sub_img_path
    target_features = utils.get_custom_target_features(sub_img)
    utils.write_custom_sig_file(sub_img_path, target_features.to_dict())
```
