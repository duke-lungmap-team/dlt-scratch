```python
border_contours = []
non_border_contours = []

for c in new_contours:
    rect = cv2.boundingRect(c)
    
    c_min_x = rect[0]
    c_min_y = rect[1]
    c_max_x = rect[0] + rect[2] - 1
    c_max_y = rect[1] + rect[3] - 1
    
    c_mins = set([c_min_x, c_min_y])
    c_maxs = set([c_max_x, c_max_y])
    
    if len(mins.intersection(c_mins)) > 0 or len(maxs.intersection(c_maxs)) > 0:
        border_contours.append(c)
    else:
        non_border_contours.append(c)
```

```python
# extract the perimiter pixels, leaving out the last pixel 
# of each side as it is included in the next side (going clockwise).
# This makes all the side arrays the same length.
# We also flip the bottom and left side, as we want to "unwrap" the 
# perimeter pixels in a clockwise fashion.
top = single_c_mask[0, :-1]
right = single_c_mask[:-1, -1]
bottom = np.flipud(single_c_mask[-1, 1:])
left = np.flipud(single_c_mask[1:, 0])
```

```python
top.shape, right.shape, bottom.shape, left.shape
```

```python
# combine the perimeter sides into one continuous array
perim_pixels = np.concatenate([top, right, bottom, left])
```

```python
region_boundary_locs = np.where(perim_pixels == 255)[0]
```

```python
region_boundary_locs
```

```python
# the perimeter here is not a geometric perimeter but the number of pixels around the image
img_h = img.shape[0]
img_w = img.shape[1]
perimeter = (img_h - 1) * 2 + (img_w - 1) * 2
perimeter
```

```python
# account for the wrap around from the last contour pixel to the end, i.e. back at the start at (0, 0)
wrap_distance = region_boundary_locs.max() - perimeter
```

```python
wrap_distance
```

```python
# insert the wrap distance in front of the region boundary locations
perim_pixels = np.concatenate([[wrap_distance], region_boundary_locs])
```

```python
perim_pixels
```

```python
# calculate the gap size between boundary pixel locations
gaps = np.diff(perim_pixels)
```

```python
# add one to the results because of the diff offset
max_gap_idx = np.where(gaps == gaps.max())[0] + 1
```

```python
# there should only be one, else we have a tie and should probably ignore that case
len(max_gap_idx)
```

```python
max_gap_idx[0]
```

```python
start_perim_loc_of_flood_entry = perim_pixels[max_gap_idx[0]]
```

```python
start_perim_loc_of_flood_entry
```

```python
# see if subsequent perimeter locations were also part of the contour, adding one to the last one we found
subsequent_region_border_locs = perim_pixels[perim_pixels > start_perim_loc_of_flood_entry]
```

```python
subsequent_region_border_locs
```

```python
floodfill_entry_point = start_perim_loc_of_flood_entry

for loc in subsequent_region_border_locs:
    if loc == floodfill_entry_point + 1:
        floodfill_entry_point = loc
```

```python
# we should hit the first interior empty point of the contour by moving forward one pixel around the perimeter
floodfill_entry_point += 1
```

```python
# now to convert our perimeter location back to an image coordinate
if floodfill_entry_point < img_w:
    print 'top'
elif floodfill_entry_point < img_w + img_h:
    print 'right'
elif floodfill_entry_point < img_w * 2 + img_h:
    print 'bottom'
else:
    print 'left'
    flood_fill_entry_coords = (0, perimeter - floodfill_entry_point)
```
