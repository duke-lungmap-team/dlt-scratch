```python
import os
import re
import glob
import cv2
import numpy as np
import pandas as pd

from lung_map_utils import utils

%matplotlib inline
np.seterr(all='raise')
```

```python
fig_size = (4, 4)
base_train_dir = "data/mouse_16.5_20x/LMEX0000000073_acta2_sftpc_sox9/poly"
class_paths = glob.glob("/".join([base_train_dir, "*"]))
```

```python
train_images = []
classes = []

for i, class_path in enumerate(sorted(class_paths)):
    if not os.path.isdir(class_path):
        continue
    folder, class_name = os.path.split(class_path)
    classes.append(class_name)

    train_images.extend(glob.glob("/".join([class_path, "*[0-9].tif"])))
```

```python
len(train_images), classes
```

```python
train_images[0]
```

```python
for sub_img_path in train_images:
    m = re.search("(.*)\.(tif)$", sub_img_path)
    g = m.groups()
    
    if len(g) > 0:
        mask_path = '{0}_{1}.tif'.format(g[0], 'mask')
    else:
        continue
    
    sub_img = cv2.imread(sub_img_path)
    mask_img = cv2.imread(
        mask_path,
        cv2.IMREAD_GRAYSCALE
    )
    
    try:
        sub_img = cv2.cvtColor(sub_img, cv2.COLOR_BGR2HSV)
        
        target_features = utils.get_custom_target_features(sub_img, mask=None, show_plots=False)
        utils.write_custom_sig_file(sub_img_path, target_features.to_dict(), suffix='custom')
        
        target_features = utils.get_custom_target_features(sub_img, mask=mask_img, show_plots=False)
        utils.write_custom_sig_file(sub_img_path, target_features.to_dict(), suffix='custom_mask')
        
        target_features = utils.get_target_features_from_tif(sub_img_path)
        utils.write_custom_sig_file(sub_img_path, target_features.to_dict(), suffix='wndcharm')
    except:
        print g[0]
```
